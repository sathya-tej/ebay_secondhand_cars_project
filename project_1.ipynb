```python
autos = autos.rename(columns={'dateCrawled':'date_Crawled','offerType':'offer_Type','vehicleType':'vehicle_Type','fuelType':'fuel_Type','nrOfPictures':'nr_Of_Pictures','postalCode':'postal_Code','lastSeen':'last_Seen'})
```

```python
autos['price'] = autos['price'].astype(str).str.replace('$', '').str.replace(',', '').astype(int)
```

```python
autos[['price','kilometer']].describe()
```

```python
# data cleaning

# we will mainly on only two variables which are price and kilometer (since more the variables more diffucult it eill become to clean the data)
# and we will use box and whiskers for cleaning data

# as from the above pricing the IQR of price and kilometer are 6050 and 25000 respictively so the optimal range for price and kilometer are:
# price = (Q1-1.5*IQR,Q3+1.5*IQR) = (0,16275) {as 0 is min so we'r etaking 0 as min}= (0,17000){rounded off}
# kilometer = (Q1-1.5*IQR,Q3+1.5*IQR) = (87500, 150000)
```

```python
autos = autos[autos["price"].between(0, 17000)]
print(autos["price"].describe())
```

```python
autos = autos[autos["kilometer"].between(87500, 150000)]
print(autos["kilometer"].describe())
```

```python
top_brands = autos["brand"].value_counts().head(10).index
print(f"Top 10 Brands: {top_brands}")
```

```python
# as we dont have the data on how much mileage it has so we will consider mileage as total km travelled assuming all have same mileage
```

```python
brand_mean_prices = {}

for brand in top_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_price = brand_only["price"].mean()
    brand_mean_prices[brand] = int(mean_price)

brand_mean_mileage = {}
for brand in top_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_mileage = brand_only["kilometer"].mean()
    brand_mean_mileage[brand] = int(mean_mileage)

bpm_series = pd.Series(brand_mean_prices)
bmm_series = pd.Series(brand_mean_mileage)

brand_info = pd.DataFrame(bpm_series, columns=['mean_price'])
brand_info['mean_mileage'] = bmm_series

brand_info['ratio'] = brand_info['mean_mileage'] / brand_info['mean_price']

print(brand_info.sort_values("ratio", ascending= False))
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=brand_info,
    x="mean_mileage",
    y="mean_price",
    s=150,
    hue=brand_info.index,
    palette="deep"
    )

for brand in brand_info.index:
    plt.text(
        brand_info.loc[brand, "mean_mileage"] + 1000, # Offset x slightly
        brand_info.loc[brand, "mean_price"] + 100,    # Offset y slightly
        brand,
        fontsize=10
        )

plt.title("Average Price vs. Average Mileage (Top 10 Brands)")
plt.xlabel("Average Mileage (km)")
plt.ylabel("Average Price ($)")
plt.show()
```
